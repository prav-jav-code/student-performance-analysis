```python
import pandas as pd

df = pd.read_csv("students.csv")
df.head()
```

```python
# Create average column
df["Average"] = df[["Maths", "Science", "English"]].mean(axis=1)

# Show updated data
df
```

```python
top_student = df.loc[df["Average"].idxmax()]
top_student
```

```python
df.mean(numeric_only=True)
```

```python
import matplotlib.pyplot as plt

plt.bar(df["Name"], df["Average"])
plt.title("Average Marks of Students")
plt.xlabel("Students")
plt.ylabel("Average Marks")
plt.show()
```

```python
df[["Maths", "Science", "English"]].mean()
```

```python
subject_avg = df[["Maths", "Science", "English"]].mean()

plt.pie(subject_avg, labels=subject_avg.index, autopct='%1.1f%%')
plt.title("Subject-wise Performance")
plt.show()
```

```python
df[["Maths", "Science", "English"]].corr()
```

```python
import matplotlib.pyplot as plt

corr = df[["Maths", "Science", "English"]].corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns)
plt.yticks(range(len(corr)), corr.columns)

for i in range(len(corr)):
    for j in range(len(corr)):
        plt.text(j, i, round(corr.iloc[i, j], 2),
                 ha="center", va="center", color="black")

plt.title("Correlation Heatmap")
plt.show()
```

```python
def category(avg):
    if avg >= 85:
        return "Top"
    elif avg >= 70:
        return "Average"
    else:
        return "Weak"

df["Category"] = df["Average"].apply(category)

df
```
